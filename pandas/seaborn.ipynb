#### Data Visualization with SeaBorn 
Seaborn is a python visualization library based on matplotlib that provides a high-level interface for drawing attractive and information statistical graphics. seaborn hr\elps in creating complex visualization with just a few lines of code.

```python
import seaborn as sns
```

```python
##Basic plotting with seaborn
tips = sns.load_dataset('tips')
tips
```

```python
##create scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of total bill vs Tip")
plt.show()
```

```python
#Line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot for Total Bill by Size")
plt.show()
```

```python
##Categorial Variable
##Bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar Plot for Total Bill By Day")
plt.show()
```

```python
#Box Plot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
##Violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
##Histogram
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
##KDE plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
##Pairplot
sns.pairplot(tips)
```

```python
##HeatMap
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
