```python
#Merging and joining dataframes
df1 = pd.DataFrame({'key' : ['A','B','C'],'value1' : [1,2,3]})
df2 = pd.DataFrame({'key' : ['A','B','D'],'value2' : [4,5,6]})
```

```python
df1
```

```python
df2
```

```python
#Merge - Inner Join
pd.merge(df1,df2,on='key',how='inner')#Only rows with the same id in both tables appear in the result. Drops everything else
```

```python
#Outer Join-  All rows from both tables.
#Matches â†’ merged into one row.
#Non-matches â†’ filled with NaN on the missing side.
pd.merge(df1,df2,on="key",how="outer")
```

```python
#Left Join -
# All rows from the left table.
#If a match exists in the right table â†’ show it.
#If not â†’ NaN in right-side columns.
pd.merge(df1,df2,on="key",how="left")
```

```python
##Right - Join
#Keeps: All rows from the right table.
#If a match exists in the left â†’ show it.
#If not â†’ NaN in left-side columns.
pd.merge(df1,df2,on="key",how="right")
```
