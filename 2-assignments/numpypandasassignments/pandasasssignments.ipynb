```python
#8.1
#A pivot table in pandas is a way to summarize and reshape data — similar to Excel’s pivot tables.
#It lets you group data by one or more keys, apply aggregations (sum, mean, count, etc.), and display the result in a matrix format.

date = pd.date_range(start='2022-01-01',end='2022-01-10',freq='D')
df = pd.DataFrame({'Date':np.random.choice(date,size=20),'Category': np.random.choice(['A','B','C'],size=20),'Value': np.random.randint(1,100,size=20)})
df

df_pivot = df.pivot_table(values='Value',index='Date',columns='Category',aggfunc='sum')
df_pivot
```

```python
year = np.random.choice(['2021','2022','2023'],size=12)
quarter = np.random.choice(['q1','q2','q3','q4'],size=12)
revenue = np.random.randint(1,100,size=12)
dataFrame = pd.DataFrame({'Year' : year,'Quarter':quarter,'Revenue':revenue})
dataFrame

dataFrame_pivot = dataFrame.pivot_table(values='Revenue',index='Year',columns='Quarter',aggfunc='mean')
dataFrame_pivot
```
